Данный рейтинг предназначен для маркировки игр для США и Канады. И логично, что для других регионов он не заполняется. А в Японии есть свой рейтинг, свой рейтинг есть в ЕС и отдельно в Германии, Австралии и т.д., но по ним в датасете нет данных.

В данных есть следующие рейтинги:  
M - Подходит для лиц старше 17 лет. Этот рейтинг используется, когда насилие становится реалистичным и похожим на реальную жизнь. Также может присутствовать ненормативная лексика, употребление табака, алкоголя или запрещенных наркотиков. Примеры: Ark: Survival Evolved, Destiny 2.  
E - Подходит для всех возрастных категорий. Игра не должна содержать звуков или изображений, которые могут напугать маленьких детей. Не должна использоваться ненормативная лексика. Примеры: Просто таннцуй, FIFA.  
T - Подходит для лиц старше 13 лет. Игра может содержать более реалистичные и натуралистические сцены насилия. Примеры: Fortnite, Sims 4.  
E 10+ - Подходит для лиц старше 10 лет. Игра может содержать умеренные формы насилия, а некоторые сцены могут пугать детей. Примеры: Minecraft Dungeons, Растения против зомби.  

Рейтинги "Е" (Everyone) и "К-А" (Kids to Adults) в контексте игр — это схожие категории, и они обозначают, что игра подходит для всех возрастов. В разные периоды времени использовались разные обозначения, но обе категории предполагают отсутствие контента, не подходящего для детей, поэтому заменим K-A на E

Для начала посмотрим за какие года в таблицы имеются данные.Исходя видим, что в 1980 году вышла самая первая игра, а в 2016 году.Напоминаю что эти даты только за период продаж в интернет-магазине «Стримчик» а не за все время.

```python
counts_year = data['year_of_release'].value_counts() #количество строк для каждого года
weight_year = (counts_year / len(data)) * 100  #выводим удельный вес каждого года
df_year_group = pd.DataFrame({'count': counts_year, 'weight_year,%': weight_year}) #создаем датафрейм с годами
df_year_group
```

```python
# сортировка данных по годам т.к df_year_group у меня идут по убыванию сумм
df_year_group_sorted = df_year_group.sort_index()

# построение столбчатой диаграммы
ax = df_year_group_sorted['count'].plot(kind='bar', figsize=(17, 13), color='skyblue', grid=False)

# добавление значений на столбцы
for i, count in enumerate(df_year_group_sorted['count']):
    ax.text(i, count + 10, str(count), ha='center', va='bottom', fontsize=10)

plt.title('Количество выпущенных игр по годам')
plt.xlabel('Год выпуска')
plt.ylabel('Количество игр')

# Увеличение размера подписей на осях
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

plt.show()
```

Создадим новую таблицу platform_grouped где отобразим сразу платформы,кол-во игры их сумму продаж, а так же удельные веса

```python
#создаем по платформе данные по кол-ву игр и суммам продаж
platform_grouped = data.groupby('platform')['total_sales'].agg(['count', 'sum']) 
platform_grouped.columns = ('Количество платформ', 'Сумма продаж игр') #переименование названий столбцов

#расчет удельного веса платформ
platform_grouped['Удельный вес по количествам платформ,%'] = platform_grouped['Количество платформ'] / len(data) * 100
#расчет удельного веса сумм продаж
platform_total_sales = platform_grouped['Сумма продаж игр'].sum() #для расчета удельного веса подсчитываем итоговую сумму продаж
platform_grouped['Удельный вес по суммам продаж,%'] = platform_grouped['Сумма продаж игр'] / platform_total_sales * 100

#сортировка по убыванию удельного веса сумм продаж
platform_grouped = platform_grouped.sort_values(by='Удельный вес по суммам продаж,%', ascending=False)

platform_grouped['Удельный вес по количествам платформ,%'] = platform_grouped['Удельный вес по количествам платформ,%'].round(2)
platform_grouped['Удельный вес по суммам продаж,%'] = platform_grouped['Удельный вес по суммам продаж,%'].round(2)

platform_grouped
```

#### Шаг 5. Проверка гипотез

**пороговое alpha значение задается самостоятельно**

**5.1.Средние пользовательские рейтинги платформ Xbox One и PC одинаковые**

Сформулируем нулевую и альтернативную гипотезы:  

**Нулевая гипотеза (H0)**: Средние пользовательские рейтинги платформ Xbox One и PC равны.

**Альтернативная гипотеза (H1)**: Средние пользовательские рейтинги платформ Xbox One и PC не равны.

```python
ratings_xbox = data_actual[data_actual['platform'] == 'XOne']['user_score'].dropna()
ratings_pc = data_actual[data_actual['platform'] == 'PC']['user_score'].dropna()

# Проведем t-тест
t_statistic, p_value = ttest_ind(ratings_xbox, ratings_pc)

# Выведем результаты
print(f"Значение t-статистики: {t_statistic}")
print(f"P-значение: {p_value}")

# Задаем уровень значимости (например, 0.05) и делаем вывод
alpha = 0.05
if p_value < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не отвергаем нулевую гипотезу")
```

```python
# средние пользовательские рейтинги по платформам
ratings_xbox = data_actual[data_actual['platform'] == 'XOne']['user_score'].dropna().mean()
ratings_pc = data_actual[data_actual['platform'] == 'PC']['user_score'].dropna().mean()
dif_pl = ratings_xbox - ratings_pc
print('Средние пользовательские рейтинги по XOne:',ratings_xbox)
print('Средние пользовательские рейтинги по PC:',ratings_pc)
print('Разница',dif_pl)
```

Исходя из постановки 'Средние пользовательские рейтинги платформ Xbox One и PC одинаковые' сформулировали гипотезу равно или не равно по  условию.  
Обычно принимают стандартное значение 0.05 его и оставили  
Выбрали ttest_ind в данном случае потому, что у нас есть две независимые выборки (рейтинги пользователей для двух разных платформ: Xbox One и PC), и мы хотим проверить, различаются ли их средние значения
По умолчанию ttest_ind считается двусторонним тестом (alternative='two-sided'), что означает, что он проверяет, различаются ли средние значения в двух выборках. Если бы мы хотели проверить, отличаются ли рейтинги пользователей на Xbox One от рейтингов на PC в какую-то конкретную сторону (например, больше на Xbox One), то можно было бы указать альтернативную гипотезу (alternative), а  в данном случае не указываем.

Полученное p-значение равно 0.14, что больше выбранного уровня значимости 0.05. Следовательно, у нас нет оснований отвергнуть нулевую гипотезу. Таким образом, можно утверждать что средние пользовательские рейтинги платформ Xbox One и PC равны.


**5.2.Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные**

Формулируем гипотезы:  
**Нулевая гипотеза (H0)**: Средние пользовательские рейтинги жанров Action и Sports равны.  
**Альтернативная гипотеза (H1)**: Средние пользовательские рейтинги жанров Action и Sports не равны.

```python
ratings_action = data_actual[data_actual['genre'] == 'Action']['user_score'].dropna()
ratings_sports = data_actual[data_actual['genre'] == 'Sports']['user_score'].dropna()

# Проведение t-теста
t_statistic, p_value = ttest_ind(ratings_action, ratings_sports)

# Вывод результатов
print(f"Значение t-статистики: {t_statistic}")
print(f"P-значение: {p_value}")

# Задаем уровень значимости (например, 0.05) и делаем вывод
alpha = 0.05
if p_value < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не отвергаем нулевую гипотезу")
```

```python
# средние пользовательские рейтинги по жанрам
ratings_action = data_actual[data_actual['genre'] == 'Action']['user_score'].dropna().mean()
ratings_sports = data_actual[data_actual['genre'] == 'Sports']['user_score'].dropna().mean()
dif_ge = ratings_action - ratings_sports
print('Средние пользовательские рейтинги по жанру Action:',ratings_action)
print('Средние пользовательские рейтинги по жанру Sports:',ratings_sports )
print('Разница',dif_ge)
```

Гипотезы сформулировали исходя из постановки заданного текста 'Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные' соответсвенно приняли равно или не равно.
ttest_ind используется для сравнения средних двух независимых выборок. В данном случае у нас есть два разных жанра (Action и Sports), и мы хотим сравнить их пользовательские рейтинги.
Если бы у нас было одно значение для сравнения (например, средний рейтинг по всем играм), мы бы использовали одновыборочный t-тест (scipy.stats.ttest_1samp).

Итак, результаты t-теста указывают на то, что существует статистически значимое различие между средними пользовательскими рейтингами жанров Action и Sports. Значение p-уровня значимости очень мало (примерно 1.05e-27), что намного меньше уровня значимости (обычно 0.05).

Таким образом, у нас есть достаточно оснований отвергнуть нулевую гипотезу о равенстве средних пользовательских рейтингов жанров Action и Sports в пользу альтернативной гипотезы о том, что они различаются. Это может подразумевать, Средние пользовательские рейтинги жанров Action и Sports не равны.

   **-Проверка гипотез.**   
Проверили две гипотезы.  

**1.Средние пользовательские рейтинги платформ Xbox One и PC одинаковые**   
Нулевая гипотеза (H0): Средние пользовательские рейтинги платформ Xbox One и PC равны.  
Альтернативная гипотеза (H1): Средние пользовательские рейтинги платформ Xbox One и PC не равны.

Полученное p-значение равно 0.14, что больше выбранного уровня значимости 0.05. Следовательно, у нас нет оснований отвергнуть нулевую гипотезу. Таким образом, можно утверждать что средние пользовательские рейтинги платформ Xbox One и PC равны.  

**2.Средние пользовательские рейтинги жанров Action и Sports разные**  
Нулевая гипотеза (H0): Средние пользовательские рейтинги жанров Action и Sports равны.  
Альтернативная гипотеза (H1): Средние пользовательские рейтинги жанров Action и Sports не равны.  

Итак, результаты t-теста указывают на то, что существует статистически значимое различие между средними пользовательскими рейтингами жанров Action и Sports. Значение p-уровня значимости очень мало (примерно 1.05e-27), что намного меньше уровня значимости (обычно 0.05).Таким образом, у нас есть достаточно оснований отвергнуть нулевую гипотезу о равенстве средних пользовательских рейтингов жанров Action и Sports в пользу альтернативной гипотезы о том, что они различаются. Это может подразумевать, Средние пользовательские рейтинги жанров Action и Sports не равны.
